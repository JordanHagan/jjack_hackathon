```python
import pandas as pd
```

```python
df = pd.read_pickle('data/dataframe.pkl')
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
df['Date'] = df['Date'].str.replace('Jan','01').str.replace('Feb','02').str.replace('Mar','03')
df['Date'] = df['Date'].str.replace('Apr','04').str.replace('May','05').str.replace('Jun','06')
df['Date'] = df['Date'].str.replace('Jul','07').str.replace('Aug','08').str.replace('Sep','09')
df['Date'] = df['Date'].str.replace('Oct','10').str.replace('Nov','11').str.replace('Dec','12')
```

```python
#Takes a long time to run
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df = df.convert_objects(convert_numeric=True)
```

```python
df = df.sort_values('Date')
```
